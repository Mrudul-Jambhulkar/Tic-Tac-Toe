# TIC TAC TOE
The tic tac toe board is mapped to the numpad

```python
def display_list(r1, r2, r3):
    for row in [r1, r2, r3]:
        print(" | ".join(row))
        print("-" * 9)

r1 = ["_", "_", "_"]
r2 = ["_", "_", "_"]
r3 = [" ", " ", " "]
display_list(r1, r2, r3)
nums=['1','2','3','4','5','6','7','8','9']

count = 0

while True:
    loc1 = input("Player 1 (X): Select a location (1-9): ")
    while loc1 not in nums:
        print('Please enter valid input')
        loc1 = input("Player 1 (X): Select a location (1-9): ")
    if loc1 == "1":
        r3[0] = "X"
    elif loc1 == "2":
        r3[1] = "X"
    elif loc1 == "3":
        r3[2] = "X"
    elif loc1 == "4":
        r2[0] = "X"
    elif loc1 == "5":
        r2[1] = "X"
    elif loc1 == "6":
        r2[2] = "X"
    elif loc1 == "7":
        r1[0] = "X"
    elif loc1 == "8":
        r1[1] = "X"
    elif loc1 == "9":
        r1[2] = "X"

    display_list(r1, r2, r3)
    count += 1

    
    if r1 == ["X", "X", "X"] or r2 == ["X", "X", "X"] or r3 == ["X", "X", "X"]:
        print("Player 1 wins")
        break
    elif r1[0] == r2[1] == r3[2] == "X" or r1[2] == r2[1] == r3[0] == "X":
        print("Player 1 wins")
        break
    elif (r1[0] == r2[0] == r3[0] == "X") or (r1[1] == r2[1] == r3[1] == "X") or (r1[2] == r2[2] == r3[2] == "X"):
        print("Player 1 wins")
        break

    if count == 9:
        print("Match tied")
        break

    loc2 = input("Player 2 (O): Select a location (1-9): ")
    while loc2 not in nums:
        print('Please enter valid input')
        loc2 = input("Player 2 (O): Select a location (1-9): ")
    if loc2 == "1":
        r3[0] = "O"
    elif loc2 == "2":
        r3[1] = "O"
    elif loc2 == "3":
        r3[2] = "O"
    elif loc2 == "4":
        r2[0] = "O"
    elif loc2 == "5":
        r2[1] = "O"
    elif loc2 == "6":
        r2[2] = "O"
    elif loc2 == "7":
        r1[0] = "O"
    elif loc2 == "8":
        r1[1] = "O"
    elif loc2 == "9":
        r1[2] = "O"
    

    display_list(r1, r2, r3)
    count += 1

    if r1 == ["O", "O", "O"] or r2 == ["O", "O", "O"] or r3 == ["O", "O", "O"]:
        print("Player 2 wins")
        break
    elif r1[0] == r2[1] == r3[2] == "O" or r1[2] == r2[1] == r3[0] == "O":
        print("Player 2 wins")
        break
    elif (r1[0] == r2[0] == r3[0] == "O") or (r1[1] == r2[1] == r3[1] == "O") or (r1[2] == r2[2] == r3[2] == "O"):
        print("Player 2 wins")
        break

    if count == 9:
        print("Match tied")
        break
```
